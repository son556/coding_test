# 18. 예산

문제 설명
S사에서는 각 부서에 필요한 물품을 지원해 주기 위해 부서별로 물품을 구매하는데 필요한 금액을 조사했습니다. 그러나, 전체 예산이 정해져 있기 때문에 모든 부서의 물품을 구매해 줄 수는 없습니다. 그래서 최대한 많은 부서의 물품을 구매해 줄 수 있도록 하려고 합니다.

물품을 구매해 줄 때는 각 부서가 신청한 금액만큼을 모두 지원해 줘야 합니다. 예를 들어 1,000원을 신청한 부서에는 정확히 1,000원을 지원해야 하며, 1,000원보다 적은 금액을 지원해 줄 수는 없습니다.

부서별로 신청한 금액이 들어있는 배열 d와 예산 budget이 매개변수로 주어질 때, 최대 몇 개의 부서에 물품을 지원할 수 있는지 return 하도록 solution 함수를 완성해주세요.

제한사항
- d는 부서별로 신청한 금액이 들어있는 배열이며, 길이(전체 부서의 개수)는 1 이상 100 이하입니다.
- d의 각 원소는 부서별로 신청한 금액을 나타내며, 부서별 신청 금액은 1 이상 100,000 이하의 자연수입니다.
- budget은 예산을 나타내며, 1 이상 10,000,000 이하의 자연수입니다.

입출력 예

d	budget	result

[1,3,2,5,4]	9	3

[2,2,3,3]	10	4

```python
# 풀이(O)
def solution(d, budget):
    answer = 0
    d.sort()
    st = 0
    x = 0
    for i in d:
        x += i
        st += 1
        if x == budget:
            answer = st
            break
            
        elif x > budget:
            if d[0] > budget:
                answer = 0
                break
            else:
                answer = st-1
                break
        
        else:
            if st == len(d):
                answer = len(d)
    return answer
```

```python
d = [1,1,2,3,5] 
budget = 10
solution(d, budget)
```

```python
# 풀이(X)
def solution(d, budget):
    answer = 0
    d.sort()
    x = 0
    idx = 0
    count = 0
    n_list = []
    while answer == 0:
        x += d[idx]
        n_list.append(d[idx])
        idx += 1
        
        if x == budget:
            answer = len(n_list) 
            
        elif x > budget:
            count += 1
            if d[0] > budget:
                answer = 0
                break
                
            if count == 1:
                st_list = n_list
            if count == len(d)-1:
                answer = len(d[0:d.index(st_list[-1])])
            x = 0
            idx = count
            n_list = []
                
        else:
            if len(n_list) == len(d):
                answer = len(n_list)
        
    return answer
```

```python
d = [1,1,2,3,5] 
budget = 10
solution(d, budget)
# 중요한건 최대로 줄수있는 인원이지 예산에 맞게주는 것이 아님..
# 나는 예산에 맞게 최대로 줘야한다고 생각, 위의 경우 예산에 맞게주면 3
# 예산에 맞지 않게 주면 4이다. 굳이 예산에 안맞는다고 시작점을 1칸씩 미뤄
# 가며 예산을 맞추는 경우를 찾을 필요 없었음..
```

### itertools 의 combinations 에서 [2,2,4,4] 중 n개를 뽑는 경우의 수 에서 리스트의 2(또는 4)들은 서로 다른 요소로 인식

```python
from itertools import combinations
n_list = [2,2,4,4]
a = list(combinations(n_list,3))
print(a)
sum(a[0])
```

### pop(요소)를 하면 해당요소를 리스트에서 꺼낸다. pop()를 하면 마지막 위치의 요소를 꺼낸다. pop(인덱스)를 하면 인덱스에 해당하는 요소를 리스트에서 꺼낸다.

```python
d = ['a','b','c','d','e'] 
print(d.pop(1))
print()
print(d)
```

### 리스트가 숫자로 이루어져 있고 pop()의 ()안에 숫자가 들어오면 ()안의 숫자는 인덱스로 인식하여 인덱스에 해당하는 요소를 리스트에서 꺼낸다.

```python
a = [4,3,2,1,0]
print(a.pop(3))
print(a)
```
