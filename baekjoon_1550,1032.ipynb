### 16진수 10진수 변환

```python
n = 'A'
dict_16 = {'A':10,'B':11,'C':12,'D':13,'E':14,'F':15}
a = len(n) - 1
num = 0
for i in n:
    if i.isdigit():
        num += int(i)*(16**a)
    else:
        num += dict_16[i]*(16**a)
    a -= 1
print(num)
```

```python
# 1032
arr = ['3','a','a','c','a']
answer = []
result = ''

if len(arr) > 2:
    for x,y in zip(arr[1],arr[2]):
        if x==y:
            answer.append(x)
        else:
            answer.append('?')
    
    for i in arr[3:]:
        n=0
        for a,b in zip(answer,i):
            if a==b:
                n+=1  
            else:
                answer[n]='?'
                n+=1
else:
    answer.append(arr[1])
print(result.join(answer))
```
