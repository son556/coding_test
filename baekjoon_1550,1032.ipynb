### 1550 16진수 10진수 변환

```python
n = 'A'
dict_16 = {'A':10,'B':11,'C':12,'D':13,'E':14,'F':15}
a = len(n) - 1
num = 0
for i in n:
    if i.isdigit():
        num += int(i)*(16**a)
    else:
        num += dict_16[i]*(16**a)
    a -= 1
print(num)
```

```python
n = int(input())
arr = []
result = ''
answer = []
for _ in range(n):
    arr.append(input())
if len(arr) > 1:
    for x,y in zip(arr[0],arr[1]):
        if x==y:
            answer.append(x)
        else:
            answer.append('?')
    
    for i in arr[2:]:
        n=0
        for a,b in zip(answer,i):
            if a==b:
                n+=1  
            else:
                answer[n]='?'
                n+=1
else:
    answer.append(arr[0])
for i in answer:
    result += i
print(result)
```

### 추가문제

```python
answer = input().split()
news = input().split()
result = []
final_result =''
num = int(answer[0])*int(answer[1])
for i in news:
    result.append(int(i)-num)
for i in result:
    final_result += str(i) + ' '
print(final_result)
```
