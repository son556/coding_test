```python
- numbers 배열의 크기는 1 이상 1,000 이하입니다.
- numbers 배열 원소의 값은 0 이상 9 이하인 정수입니다.
hand는 "left" 또는 "right" 입니다.
"left"는 왼손잡이, "right"는 오른손잡이를 의미합니다.
왼손 엄지손가락을 사용한 경우는 L, 
오른손 엄지손가락을 사용한 경우는 R을 순서대로 이어붙여 
문자열 형태로 return 해주세요.
입출력 예
numbers	hand	result
[1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5]	"right"	"LRLLLRLLRRL"
[7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2]	"left"	"LRLLRRLLLRR"
[1, 2, 3, 4, 5, 6, 7, 8, 9, 0]	"right"	"LLRLLRLLRL"
left -> [1,4,7]
right -> [3,6,9]
[2,5,8,0]->'판단'
```

```python
# 처음풀이
def solution(numbers, hand):
    answer = ''
    handic = {'R':'#','L':'*'}
    dict2 = {'1':['1','3','5'], '2':['4','6','8'],'3':['0','7','9'],
           '4':['*','#']}
    dict5 = {'1':['2','4','6','8'], '2':['1','3','7','9','0'],
            '3':['*','#']}
    dict8 = {'1':['5','7','9','0'], '2':['2','4','6','*','#'],
            '3':['1','3']}
    dict0 = {'1':['*','#','8'], '2':['5','7','9',],
            '3':['2','4','6'], '4':['1','3']}
    n_list = ''
    number = '2580'
    for i in numbers:
        n_list += str(i)
    for idx, value in enumerate(n_list):
        if value in '147':
            answer += 'L'
            handic['L'] = value
        
        elif value in '369':
            answer += 'R'
            handic['R']= value
            
        if hand == 'right':   
            if value == '2':
                for j in dict2:
                    for i in handic:
                        if handic['R'] in dict2[j] and handic['L'] in dict2[j]:
                            handic['R'] = '2'
                            answer += 'R'
                            break
                        elif handic[i] in dict2[j]:
                            handic[i] = '2'
                            answer += i
                            break
                    if handic['R'] == '2' or handic['L'] =='2':
                            break
            
            elif value == '5':
                for j in dict5:
                    print('j:',j)
                    for i in handic:
                        print('i:',i)
                        if handic['R'] in dict5[j] and handic['L'] in dict5[j]:
                            print('a')
                            handic['R'] = '5'
                            answer += 'R'
                            break
                        elif handic[i] in dict5[j]:
                            print('b')
                            handic[i] = '5'
                            answer += i
                            break
                    if handic['R'] == '5' or handic['L'] =='5':
                        break
                        
            
            elif value == '8':
                for j in dict8:
                    for i in handic:
                        if handic['R'] in dict8[j] and handic['L'] in dict8[j]:
                            handic['R'] = '8'
                            answer += 'R'
                            break
                        elif handic[i] in dict8[j]:
                            handic[i] = '8'
                            answer += i
                            break  
                    if handic['R'] == '8' or handic['L'] =='8':
                        break
            
            elif value == '0':
                for j in dict0:
                    for i in handic:
                        if handic['R'] in dict0[j] and handic['L'] in dict0[j]:
                            handic['R'] = '0'
                            answer += 'R'
                            break
                        elif handic[i] in dict0[j]:
                            handic[i] = '0'
                            answer += i
                            break
                    if handic['R'] == '0' or handic['L'] =='0':
                        break
    
    
    return answer, handic
```

```python
# 풀이 줄이기 -> 실패(8,12,15,16,18,19,20)
def solution(numbers, hand):
    answer = ''
    handic = {'R':'#','L':'*'}
    dict2 = {'1':['1','3','5'], '2':['4','6','8'],'3':['0','7','9'],
           '4':['*','#']}
    dict5 = {'1':['2','4','6','8'], '2':['1','3','7','9','0'],
            '3':['*','#']}
    dict8 = {'1':['5','7','9','0'], '2':['2','4','6','*','#'],
            '3':['1','3']}
    dict0 = {'1':['*','#','8'], '2':['5','7','9',],
            '3':['2','4','6'], '4':['1','3']}
    n_list = ''
    number = {'2':dict2, '5':dict5, '8':dict8, '0':dict0}
    for i in numbers:
        n_list += str(i)
    for idx, value in enumerate(n_list):
        if value in '147':
            answer += 'L'
            handic['L'] = value
        
        elif value in '369':
            answer += 'R'
            handic['R']= value
        
        else:    
            if hand == 'right':
                for i in number:
                    if value == i:
                        for x in number[i]:
                            for j in handic:
                                if handic[j] == i:
                                    handic[j] = i
                                    answer += j
                                elif handic['R'] in number[i][x] and\
                                handic['L'] in number[i][x]:
                                    print(handic)
                                    handic['R'] = i
                                    answer += 'R'
                                    print(i)
                                    print('{},{}'.format(handic,number[i][x]))
                                    print('----------------------')
                                elif handic[j] in number[i][x]:
                                    print(handic)
                                    handic[j] = i
                                    answer += j
                                    print(i)
                                    print('{},{}'.format(handic,number[i][x]))
                                    print('----------------------')
                            if handic['R'] == i or handic['L'] == i:
                                break
            elif hand == 'left':
                for i in number:
                    if value == i:
                        for x in number[i]:
                            for j in reversed(handic):
                                if handic[j] == i:
                                    handic[j] = i
                                    answer += j
                                    break
                                elif handic['R'] in number[i][x] and\
                                handic['L'] in number[i][x]:
                                    handic['L'] = i
                                    answer += 'L'
                                    break
                                elif handic[j] in number[i][x]:
                                    handic[j] = i
                                    answer += j
                                    break
                            if handic['R'] == i or handic['L'] == i:
                                break              
    return answer, handic
```

```python
# 다른사람풀이
def findDistance(handNumber, nextNumber):
    keypad = {1: [0, 0], 2: [1, 0], 3: [2, 0],
              4: [0, 1], 5: [1, 1], 6: [2, 1],
              7: [0, 2], 8: [1, 2], 9: [2, 2],
              "*": [0, 3], 0: [1, 3], "#": [2, 3], }

    x1, y1 = keypad[handNumber]
    x2, y2 = keypad[nextNumber]

    return abs(x1-x2) + abs(y1-y2)


def solution(numbers, hand):
    answer = ''
    leftLocation = "*"
    rightLocation = "#"

    for number in numbers:
        if number == 1 or number == 4 or number == 7:
            answer += "L"
            leftLocation = number
        elif number == 3 or number == 6 or number == 9:
            answer += "R"
            rightLocation = number
        elif number == 2 or number == 5 or number == 8 or number == 0:  
            # 키패드: 2 5 8 0
            leftDistance = findDistance(leftLocation, number)
            rightDistance = findDistance(rightLocation, number)

            if leftDistance < rightDistance:
                answer += "L"
                leftLocation = number
            elif rightDistance < leftDistance:
                answer += "R"
                rightLocation = number
            else:
                if hand == "left":
                    answer += "L"
                    leftLocation = number
                else:
                    answer += "R"
                    rightLocation = number

    return answer
```
