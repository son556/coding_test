```python
mf_1 = ['WBWBWBWB','BWBWBWBW','WBWBWBWB','BWBWBWBW','WBWBWBWB','BWBWBWBW',
           'WBWBWBWB','BWBWBWBW',] #패턴 1
mf_2 = ['BWBWBWBW','WBWBWBWB','BWBWBWBW','WBWBWBWB','BWBWBWBW','WBWBWBWB',
       'BWBWBWBW','WBWBWBWB'] # 패턴 2


class Filter:# 패턴과 일치하는 부분을 리턴해주는 필터
    def f_1(self, sample):
        self.sample = sample
        answer_list = []
        for x,y in zip(self.sample,mf_1):
            answer = ''
            for a,b in zip(x,y):
                if a == b:
                    answer += a
                else:
                    continue
            answer_list.append(answer)
        return answer_list

        

    def f_2(self, sample):
        self.sample = sample
        answer_list = []
        for x,y in zip(self.sample,mf_2):
            answer = ''
            for a,b in zip(x,y):
                if a == b:
                    answer += a
                else:
                    continue
            answer_list.append(answer)
        return answer_list

cnt = 0
idx = 0 # n
idx_1 = 0 # m
answer_list = []
while 1:# 8 by 8 단위로 이동하며 패턴과 일치하지 않는 부분의 수 리턴
    arr_2 = []
    for i in arr[idx:idx+8]:
        arr_2.append(i[idx_1:idx_1+8])

    x = Filter()
    result1 = 0
    resutl2 = 0
    count = 0
    for i in x.f_1(arr_2): #패턴 1과 비교 했을때 일치하지 않는 부분의 개수
        count += 8 - len(i)
    result1 = count
    
    count = 0
    for i in x.f_2(arr_2): #패턴 2와 비교 했을때 일치하지 않는 부분의 개수
        count += 8 - len(i)
    result2 = count
    
    answer_list.append(min(result1,result2)) #둘 중 더 적은 값을 리스트에 추가
    if idx > n-8:
        break
    if idx_1 == m-8:
        idx += 1
        idx_1 = 0
    idx_1 += 1
    
print(min(answer_list)) #리스트에 추가된 값들중 가장 작은값을 출력
```

```python
dice = [a,b,c]
x = dice.count(a)
y = dice.count(b)
z = dice.count(c)
dice_dict = {}
dice_dict[a] = x
dice_dict[b] = y
dice_dict[c] = z
result = 0
for k in dice_dict:
    if dice_dict[k] == 2:
        result = int(k)
        print(1000 + 100*result)
    elif dice_dict[k] == 3:
        result = int(k)
        print(10000 + 1000*result)
if result == 0:
    result = max(int(a),int(b),int(c))
    print(result*100)
```
