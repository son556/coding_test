### 1712 손익분기점

```python
1000 70 170 # 11
3 2 1 # -1
2100000000 9 10 # 2100000001
```

```python
a, b, c = list(map(int,input().split()))
```

```python
import math
if b>=c:
    print(-1)
else:
    answer = a/(c-b)
    if answer%1==0:
        print(int(answer+1))
    else:
        print(math.ceil(answer))
```
