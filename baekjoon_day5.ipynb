```python
n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int,input().split())))
```

```python
from itertools import combinations
num = [x for x in range(n)]
choice_list = list(combinations(num,n//2))
answer_list = []
for i in choice_list:
    result_1 = list(combinations(i,2))
    result_2 = list(combinations(set(num)-set(i),2))
    total_1 = 0
    for j in result_1:
        x, y = j
        total_1 += arr[x][y] + arr[y][x]
    
    total_2 = 0
    for j in result_2:
        x, y = j
        total_2 += arr[x][y] + arr[y][x]
    
    answer_list.append(abs(total_1 - total_2))
print(min(answer_list))
```
