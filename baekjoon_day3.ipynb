### 1110 더하기 사이클

```python
1. 26 #4
2. 0 #1
3. 1 #60
```

```python
n = input()
count = 0
if len(n)==1:
    n = '0'+n
a = n[0]
b = n[-1]

while 1:
    count += 1
    z = int(a) + int(b)
    z = str(z)
    z = b+z[-1]
    if z == n:
        print(count)
        break
    else:
        if len(z) == 1:
            z = '0'+z
        a = z[0]
        b = z[-1]
```

### 1010 다리놓기

```python
3       # 1 5 67863915
2 2
1 5
13 29
```

```python
import math
n = input()
n = int(n)
for _ in range(n):
    a, b = map(int,input().split())
    answer = math.factorial(b)/(math.factorial(a)*math.factorial(b-a))
    print(int(answer))
```

### 1145 적어도 대부분의 배수

```python
30 42 70 35 90 # 210
1 2 3 4 5 # 4
30 45 23 26 56 # 1170
3 14 15 92 65 # 195
```

```python
from itertools import combinations
arr = input().split()
for i,v in enumerate(arr):
    arr[i] = int(v)
arr.sort()

arr_2 = list(combinations(arr,3))
answer_list = []
for i in arr_2:
    a,b,c = i[0],i[1],i[2]
    for i in range(c,a*b*c+1):
        if i%a==0 and i%b==0 and i%c==0:
            answer_list.append(i)
            break
print(min(answer_list))
```

### 1157 단어공부

```python
Mississipi # ?
zZa # Z
z # Z
baaa # A
```

```python
from collections import Counter
a = input().lower()
result = Counter(a)
num = max(result.values())

for i in result:
    if num == result[i]:
        n = i
        break
for i,v in result.items():
    answer = result[i]
    if answer==result[n] and n!=i:
        answer = '?'
        break
if answer != '?':
    answer = n.upper()
print(answer)
```

### 10819 차이를 최대로

```python
6 #62
20 1 15 8 4 10
```

```python
from itertools import permutations 
N = input()
arr = list(map(int,input().split()))
arr2 = list(permutations(arr,int(N)))
result = []
answer = 0
for i in arr2:
    r = 0
    for x in range(1,int(N)):
        r += abs(i[x-1]-i[x])
    result.append(r)
answer = max(result)
print(answer)
```

### 2798 블랙잭

```python
5 21 #21
5 6 7 8 9

10 500  # 497
93 181 245 214 315 36 185 138 216 295
```

```python
from itertools import combinations
N,M = input().split()
arr = list(map(int,input().split()))
arr_2 = list(combinations(arr,3))
answer_dict = {}
for i,v in enumerate(arr_2):
    answer_dict[v[0]+v[1]+v[2]] = v[0]+v[1]+v[2]

for i in range(int(M),1,-1):
    try:
        print(answer_dict[i])
        break
    except:
        continue
```
