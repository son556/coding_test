```python
수많은 마라톤 선수들이 마라톤에 참여하였습니다. 단 한 명의 선수를 제외하고는 모든 선수가 마라톤을 완주하였습니다.

마라톤에 참여한 선수들의 이름이 담긴 배열 participant와 완주한 선수들의 이름이 담긴 배열 completion이 주어질 때, 완주하지 못한 선수의 이름을 return 하도록 solution 함수를 작성해주세요.

제한사항
마라톤 경기에 참여한 선수의 수는 1명 이상 100,000명 이하입니다.
completion의 길이는 participant의 길이보다 1 작습니다.
참가자의 이름은 1개 이상 20개 이하의 알파벳 소문자로 이루어져 있습니다.
참가자 중에는 동명이인이 있을 수 있습니다.
입출력 예
participant	completion	return
["leo", "kiki", "eden"]	["eden", "kiki"]	"leo"
["marina", "josipa", "nikola", "vinko", "filipa"]	["josipa", "filipa", "marina", "nikola"]	"vinko"
["mislav", "stanko", "mislav", "ana"]	["stanko", "ana", "mislav"]	"mislav"
```

```python
def solution(participant, completion):
    answer = ''
    participant.sort()
    completion.sort()
    n_list = participant.copy()
    for i,v in enumerate(completion):
        if participant[i] != v:
            answer = participant[i]
            break
        else:
            n_list.remove(v)
    if len(n_list) == 1:
        answer = n_list[0]
    return answer
```

```python
participant = ["mislav", "stanko", "mislav", "ana"]
completion = ["stanko", "ana", "mislav"]
solution(participant, completion)
```

```python
# 2. hash 함수
# Hash Table: 키(Key)에 데이터(Value)를 저장하는 데이터 구조
# python -> dictionary {'key':'value'}
# 장점) 데이터 저장과 읽는 속도가 빠르다, 중복확인이 쉽다.
# 단점) 일반적으로 저장공간이 좀더 많이 필요하다.
#       여러 키에 해당하는 주소가 동일할 경우 
#################################충돌을 해결하기 위한 별도 자료구조가 필요함
```

```python
from collections import Counter

def solution(participant, completion):
    cnt = Counter()
    for p in participant:
        cnt[p] += 1
        print(cnt)
        print()
    for c in completion:
        cnt[c] -= 1
        print(cnt)
        print()
    return list(x for x,y in cnt.items() if y == 1).pop();
# .pop()의 의미: 리스트의 제일 끝 값을 지우고 지운 값을 반환
```

```python
participant =['a','b','c','a','a','b']
completion = ['a','a','b','b','c']
solution(participant, completion)
```
