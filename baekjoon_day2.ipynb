### 1297 TV크기

```python
1. 52 9 16 # 25 45
2. 7 2 3 # 3 5
3. 13 7 10 # 7 10
4. 7 32 47 # 3 5
5. 11 15 16 # 7 8
# 소수점으로 나오는 경우 버림
```

```python
d,h,w = map(int,input().split())
x = (d**2/(h**2+w**2))**0.5
H = int(h*x)
W = int(w*x)
print(H, W)
```

### 1330 두 수 비교하기

```python
1. 10 2 # >
2. 5 5  # ==
```

```python
a, b = map(int,input().split())
if a>b:
    print('>')
elif a<b:
    print('<')
else:
    print('==')
```

### 1009 분산처리

```python
5             # 1 7 6 1 9
1 6
3 7
6 2
7 100
9 635
```

```python
n = int(input())
answer_dict = {2:[2,4,8,6], 3:[3,9,7,1], 7:[7,9,3,1], 8:[8,4,2,6], 9:[9,1], 4:[4,6]}
for _ in range(n):
    a,b = map(int,input().split())
    if b == 0:
        print(1)
    elif a%10 == 1 or a%10 == 5 or a%10== 6 or a%10==0:
        if a%10 != 0:
            print(a%10)
        else:
            print(10)
    elif a%10 == 2 or a%10==3 or a%10==7 or a%10==8:
        if b%4 != 0:
            print(answer_dict[a%10][b%4-1])
        else:
            print(answer_dict[a%10][3])
    elif a%10 == 4 or a%10 == 9:
        if b%2 == 1:
            print(answer_dict[a%10][0])
        else:
            print(answer_dict[a%10][1])
```
